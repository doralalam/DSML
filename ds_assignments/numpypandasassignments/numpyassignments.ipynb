### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
array = np.random.randint(1,100,size=(5,5))
print('Original Array:\n', array)

mean = np.mean(array)
median = np.median(array)
std_div = np.std(array)
var = np.var(array)
print('Mean: ',mean)
print('Median: ', median)
print('Standard Deviation: ', std_div)
print('Variance: ', var)
```

2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
## To normalize the data, we must calculate mean and std_div

array = np.random.randint(1,9,size=(3,3))
print('Original Array')
print(array)

mean = np.mean(array)
std_div = np.std(array)

normalized_data = (array-mean)/std_div

print()
print('Normalized Data')
print(normalized_data)
```

### Assignment 5: Broadcasting

1. Create a NumPy array of shape (3, 3) filled with random integers. Add a 1D array of shape (3,) to each row of the 2D array using broadcasting.

```python
array = np.random.randint(1,10,size=(3,3))
row_array = np.random.randint(1,10,size = (3,))
print('Orignal Array')
print(array)
print()
print('Row Array')
print(row_array)
print()

## Broadcasting is the vector operation when both the vectors are different in size
print('Result:')
print(array+row_array)
## Here, row array is added to the original array for each each row individually, which is known as 'Broadcasting'
```

2. Create a NumPy array of shape (4, 4) filled with random integers. Subtract a 1D array of shape (4,) from each column of the 2D array using broadcasting.

```python
array = np.random.randint(1,17,size=(4,4))
column_array = np.random.randint(1,17,size=(4,))
print('Original Array: ')
print(array)
print()
print('Column Array')
print(column_array)
print()

## Result using 1st Approach
print('Result')
print(array-column_array[:,np.newaxis])
print()

## Result using 2nd Approach
column_array_2D = column_array.reshape(4,1)
print('Column Array in 2D')
print(column_array_2D)
print()
print('Result')
print(array-column_array_2D)
```

2. Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

1. Euclidean distance between 2 points is the straight line distance given by Pythagoros

2. Formula to calculate this is:

        For 2D points A(x1,y1) and B(x2,y2)

            d(A,B) = Sqrt(((x2-x1)**2)+((y2-y2)**2))

    For a pair wise euclidean distance for a 3 point matrix A(x1,y1), B(x2,y2), C(x3,y3)

        | d(A,A)  d(A,B)  d(A,C) |
        | d(B,A)  d(B,B)  d(B,C) |
        | d(C,A)  d(C,B)  d(C,C) |

```python
data_type = [('x','i4'),('y','i4')]
array = np.array([(2,5), (3,4), (5,4)], dtype=data_type)
print('Original Array: ')
print(array)
print()

## Formula to calculate Euclidean distances
euclidean_distance = np.sqrt((array['x'][:, np.newaxis]-array['x'])**2+(array['y'][:,np.newaxis]-array['y'])**2)
print('Distances: ')
print(euclidean_distance)

```

### Assignment 10: Masked Arrays

1. Create a masked array of shape (4, 4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

```python
import numpy.ma as ma
```

```python
array = np.random.randint(1,17, size=(4,4))
print('Original Array: ')
print(array)
print()

masked_array = ma.masked_greater(array, 10)
print('Masked Array: ')
print(masked_array)
print()

sum_unmasked = masked_array.sum()
print('Sum of unmasked elements : ', sum_unmasked)
```

2. Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

```python
array = np.random.randint(1,10,size = (3,3))
print('Original Array: ')
print(array)
print()

masked_array = ma.masked_array(array, np.eye(3))
print('Diagonally Masked Array: ')
print(masked_array)
print()

mean_unmasked = masked_array.mean()
print('Mean of the unmasked elements: ',mean_unmasked)
print()
masked_array = masked_array.filled(mean_unmasked)
print('New Masked Array: ')
print(masked_array)
```
