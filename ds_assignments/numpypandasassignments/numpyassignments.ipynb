2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Replace the diagonal elements with 0.

```python
## To create an array of 4X4 size with random integers from 1 to 16
array2 = np.random.randint(1,17, size=(4,4))
print(array2)
print()

## Approach 1
np.fill_diagonal(array2,0)
print('Array with replaced diagonal elements with 0:\n',array2)
print()

## Approach 2
for n in range(0,4):
    array2[n,n]=0
print('Array with 2nd approach\n', array2)
```


2. Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
array = np.random.randint(1,26,size=(5,5))
print('5X5 Array: ')
print(array)

## To extract the elements on the border
new_array = np.concatenate((array[0,:], array[-1,:], array[1:-1,0], array[1:-1,-1]))
## first_row+last_row+first_column+last_column
print()
print('Border Elements:')
print(new_array)
```

### Assignment 3: Array Operations

1. Create two NumPy arrays of shape (3, 4) filled with random integers. Perform element-wise addition, subtraction, multiplication, and division.

```python
array1 = np.random.randint(1,21,size=(3,4))
array2 = np.random.randint(1,21,size=(3,4))
print('Array 1:')
print(array1)
print()
print('Array2')
print(array2)
print()

## Element Wise Operations i.e., vector operations
print('Vector Sum: \n', array1+array2)
print()
print('Vector Sub: \n', array1-array2)
print()
print('Vector Mul: \n', array1*array2)
print()
print('Vector Div: \n', np.round(array1/array2,1))
## Rounded of nearest 1 decimal value
print()
```


2. Create a NumPy array of shape (4, 4) with values from 1 to 16. Compute the row-wise and column-wise sum.

```python
array = np.random.randint(1,17,size=(4,4))
print(array)

row_sum  = np.sum(array, axis=1)
column_sum = np.sum(array)
print()
print('Row-wise elements sum: ', row_sum)
print()
print('Column-wise elements sum: ', column_sum)
```

### Assignment 4: Statistical Operations

1. Create a NumPy array of shape (5, 5) filled with random integers. Compute the mean, median, standard deviation, and variance of the array.

```python
array = np.random.randint(1,100,size=(5,5))
print('Original Array:\n', array)

mean = np.mean(array)
median = np.median(array)
std_div = np.std(array)
var = np.var(array)
print('Mean: ',mean)
print('Median: ', median)
print('Standard Deviation: ', std_div)
print('Variance: ', var)
```

2. Create a NumPy array of shape (3, 3) with values from 1 to 9. Normalize the array (i.e., scale the values to have a mean of 0 and a standard deviation of 1).

```python
## To normalize the data, we must calculate mean and std_div

array = np.random.randint(1,9,size=(3,3))
print('Original Array')
print(array)

mean = np.mean(array)
std_div = np.std(array)

normalized_data = (array-mean)/std_div

print()
print('Normalized Data')
print(normalized_data)
```
