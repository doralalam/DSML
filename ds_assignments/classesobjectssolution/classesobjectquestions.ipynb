Create a class named `Car` with attributes `make`, `model`, and `year`. Create an object of the class and print its attributes.

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year

ob = Car('Toyota', 'Fortuner', 2026)
print('Make: ', ob.make)
print('Model: ', ob.model)
print('Year: ', ob.year)
```

### Assignment 2: Methods in Class

Add a method named `start_engine` to the `Car` class that prints a message when the engine starts. Create an object of the class and call the method.

```python
class Car:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self, status):
        self.status = status
        if self.status == "Started":
            print('Engine Started!!!')
        else:
            print('Engine Stopped!!!')

ob = Car('Toyota', 'Fortuner', 2026)
print('Make: ', ob.make)
print('Model: ', ob.model)
print('Year: ', ob.year)
ob.start_engine('Started')
ob.start_engine('Stopped')
```

### Assignment 4: Class with Private Attributes

Create a class named `BankAccount` with private attributes `account_number` and `balance`. Add methods to deposit and withdraw money, and to check the balance. Create an object of the class and perform some operations.


```python
class BankAccount:
    def __init__(self,account_number,balance):
        self.__account_number = account_number
        self.__balance = balance
        

    def deposit(self,deposit_amount):
        self.deposit_amount = deposit_amount
        print(f'Amount {self.deposit_amount} deposited successfully!!!')
        print(f'Updated Balance is {self.deposit_amount+self.__balance}')

    def withdraw(self,withdraw_amount):
        self.withdraw_amount = withdraw_amount
        print(f'Amount {self.withdraw_amount} withdrawn successfully!!!')
        print(f'Updated Balance is {self.__balance-self.withdraw_amount}')

    def check_balance(self):
        print(f'Your latest balance is {self.__balance}')

obj = BankAccount(1234,100)
obj.deposit(100)
obj.withdraw(50)
obj.check_balance()
```
