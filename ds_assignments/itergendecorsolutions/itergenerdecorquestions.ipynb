### Assignment 1: Custom Iterator

Create a custom iterator class named `Countdown` that takes a number and counts down to zero. Implement the `__iter__` and `__next__` methods. Test the iterator by using it in a for loop.

```python
class Countdown:
    ## takes a number
    def __init__(self, value):
        self.value = value

    ## Required method for the iterator protocol - allows the object to work with for loops and iter() function
    def __iter__(self):
        return self
    
    ## core logic for iteration
    def __next__(self):
        if self.value<=0:
            raise StopIteration
        else:
            self.value = self.value-1
            return self.value
        

## create object
countdown = Countdown(10)

## make it iterable
iterator = iter(countdown)
```

```python
## for the next() sequence
print(next(iterator))
```

```python
## iterator in for loop
for val in iterator:
    print(val)
```

### Assignment 2: Custom Iterable Class

Create a class named `MyRange` that mimics the behavior of the built-in `range` function. Implement the `__iter__` and `__next__` methods. Test the class by using it in a for loop.


```python
class MyRange:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __iter__(self):
        return self
    
    def __next__(self):
        self.current = self.start
        if self.current>=self.end:
            raise StopIteration
        else:
            self.start = self.start+1
            return self.current

myrange = MyRange(1,10)
iterator = iter(myrange)
```

```python
for val in iterator:
    print(val)
```
