#### Logging with Multiple Loggers

    We can create multiple loggers for different parts of our application

```python
## To import logging

import logging
```

```python
## Create logger for Module1
logger1 = logging.getLogger('Module1')
logger1.setLevel(logging.DEBUG)

## Create logger for Module2
logger2 = logging.getLogger('Module2')
logger2.setLevel(logging.WARNING)

## Create logger for root
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
## Log messages for Module1
logger1.debug('This message is from Module1 for debug')
logger1.warning('This message is from Module1 for warning')

## Log messages for Module2
logger2.warning('This message is from Module2 for warning')
logger2.error('This message is from Module2 for error')

## Log message for Module2 at DEBUG severity which is not covered, because module2 was logged from WARNING severity level which doesn't cover DEBUG
logger2.debug('This message is from Module2 for debug')

## Log message for root level
logging.debug('This message is from basicConfig for debug')

## In the output, we can see that Module2 logger is not giving DEBUG message and names (ModuleNames and Root) are changing accordingly
```
