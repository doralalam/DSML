#### Data Visualization with Seaborn

1. Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics

2. Seaborn helps in creating complex visualizations with just a few lines of codes

```python
import seaborn as sns

## To load the sample data 'tips.csv' which is located on online repositor for seaborn
tips = sns.load_dataset('tips')
tips
```

#### Scatter Plot

```python
## To create a scatter plot 

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot for Tip against Total_Bill')
## This is a scatterplot with x label indicating the total_bill and y_label indicating the tip from the 'tips.csv' dataset
```

#### Line Plot

```python
## To create a line plot for total_bill against size

sns.lineplot(x='size', y='total_bill', data=tips)
plt.title('Line Plot with distribution for Total_Bill against Size (No. of People)')
plt.xlabel('No. of People for each bill')
plt.ylabel('Total_Bill')
```

#### Categorical Plot

In the 'tips.csv' file, sex, smoker, day and time are categorical variables

```python
## Bar Plot using Categorical Data

sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot using Categorical Data (Day on X axis)')
```

```python
##  Box Plot using Categorical Data

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot using Categorical Data (Day on X-axis)')
## In this plot, we can see data at 0th, 25th, 50th, 75th and 100th percentiles
```

```python
## Violin Plot using Categorical Data

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot using Categorical Data (Day on X-axis)')
## This is similar to Box plot, but here, we can see the distribution and mean of the data as well
```

#### Histograms

1. Histograms are used to represent the distribution of the dataset

2. Histograms divide the data into bins and then count the data points in each relative bin

```python
## To create the Histogram

sns.histplot(tips['total_bill'], bins=10, kde=True)
## kde --> Kernal Density Estimation
## It gives the smooth continuous curve that estimates the underlying probabilistic density
plt.title('Histogram for Total_Bill')
```

```python
## KDE Plot

sns.kdeplot(tips['total_bill'], fill= True)
## In older versions of seaborn shade is used instead of fill
plt.title('KDE Plot for Total_Bill')
## KDE --> Kernel Density Estimation
```

#### Pair Plot

1. Pair Plot gives graphical representation of all the numberical columns against each other

2. For example: In the 'tips' dataset, total_bill, tip and size are the numerical columns. So the pairplot will give graphs for all of them against each other as shown

3. Categorical variables are not taken into consideration

```python
## Pair Plot

sns.pairplot(tips)
```

#### Correlation

1. Tendancy of change in variable 'B' with respect to change in variable 'A'

2. Correlation was given for numerical values only. i.e., as per the 'tips' dataset, only total_bill, tip and size were considered for calculating correlation

```python
## To calculate the correlation

corr = tips[['total_bill', 'tip', 'size']].corr()
corr
## The table indicates, for a unit increment in total_bill, tip is being increased by 0.675734 (positively)
```

```python
## To create a HeatMap for the correlation

sns.heatmap(corr, annot=True, cmap='coolwarm')
## annot will give the numerical annotations inside the boxxes
## cmap will allow us to select the custom colors
plt.title('Heat Map for Tips Data')
```

```python
## To handle Energy_Production_Data and creating bar plots

import pandas as pd

production_data_df = pd.read_csv('Energy_Production_Dataset.csv')
production_data_df.head(5)
```

```python
## Bar Plot for Monthly Energy Production

plt.figure(figsize=(15,5))
sns.barplot(x='Month_Name', y='Production', data = production_data_df, estimator=sum)
## Like aggregate sum operation in matplotlib, we can directly perform that operation in seaborn using estimator which makes the code simple
plt.xlabel('Month')
plt.ylabel('Energy Produced')
plt.title('Monthly Energy Production Data')
```

```python
## Bar Plot for Seasonal Energy Production

plt.figure(figsize=(15,5))
sns.barplot(x='Season', y='Production', data = production_data_df, estimator=sum, color="green")
plt.title('Energy Produced by Season')
```
