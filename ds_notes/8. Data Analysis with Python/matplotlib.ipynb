#### Line Plot

```python
## Sample line plot

x = [1,2,3,4,5]
y = [2,4,8,11,13]

plt.plot(x,y)

## To add lables to the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.show()
```

#### plt.show() 

0. tells Matplotlib to render and display all the figures you have created so far in a window (or in the notebook output).
​

1. In normal Python scripts or terminals, plots stay in memory until you call plt.show(), which starts Matplotlib’s event loop and actually opens the figure window.
​

2. It also usually blocks further code execution until you close the plot window, letting you look at and interact with the graph.
​

3. In Jupyter notebooks, plots often appear automatically because the environment runs in interactive mode, but plt.show() is still useful to control when plots are displayed or when working with multiple figures.
​



```python
## To create a customized line plot

## Sample Data
x = [1,2,3,4,5]
y = [2,4,8,11,13]

plt.plot(x,y,color='red', linestyle='-.', marker = 'o', linewidth = 2, markersize=6)
## color is to select the color and is optional
## linestyle is to select line style and is optional
## marker is to indicate the dots at the exact co-ordinates and is optional
## linewidth is to control the thickness and is optional
## markersize is to control the thickness and is optional
plt.grid(True)
```

```python
## Multiple plots

## Sample data

x = [1,2,3,4,5]
y1 = [3,5,7,8,9]
y2 = [2,4,5,6,8]

## To set the total figure sie i.e., width, length
plt.figure(figsize=(9,5))

## 2 rows, 2 columns, 1st plot
plt.subplot(2,2,1)
plt.plot(x,y1,color='red')
plt.title('Plot-1')

plt.subplot(2,2,2)
plt.plot(x,y2,color='green')
plt.title('Plot-2')

plt.subplot(2,2,3)
plt.plot(y1,x,color='blue')
plt.title('Plot-3')

plt.subplot(2,2,4)
plt.plot(y2,x,color='orange')
plt.title('Plot-4')
```

#### Bar Plot

```python
## Bar Plot

categories = ['A', 'B', 'C', 'D', 'E']
values = [3,5,7,3,5]

plt.bar(categories, values, color = 'purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Sample Bar Chart')
plt.show()
```

#### Histograms

1. Histograms are used to represent the distribution of the dataset

2. Histograms divide the data into bins and then count the data points in each relative bin

```python
## Histogram data

x = [1,2,2,3,3,4,4,4,5,5,5,5,3,3,3]

plt.hist(x, bins=5, color='yellow', edgecolor='black')
plt.xlabel('Bins')
plt.ylabel('No. of Data Points')
plt.title('Sample Histogram Data')
```

#### Pie Chart

```python
## Sample data for pie chart

labels = ['A', 'B', 'C', 'D']
data = [20,15,23,30]

plt.pie(data,labels=labels)
```

```python
## To add custom colors and shadow effect

## Sample data for pie chart

labels = ['A', 'B', 'C', 'D']
data = [20,15,23,30]
colors = ['blue', 'green', 'red', 'purple']

plt.pie(data,labels=labels, colors=colors, shadow=True)
```

```python
## To add the percentage points to the pie chart

## Sample data for pie chart

labels = ['A', 'B', 'C', 'D']
data = [20,15,23,30]
colors = ['blue', 'green', 'red', 'purple']

plt.pie(data,labels=labels, colors=colors, shadow=True, autopct="%1.1f%%")
## 1.1f indicates the percentage upto 1 decimal point
```

```python
## To highligh a particular section, we should use explode

labels = ['A', 'B', 'C', 'D']
data = [20,15,23,30]
colors = ['blue', 'green', 'red', 'purple']
explode = [0.2, 0, 0, 0]

plt.pie(data,labels=labels, colors=colors, shadow=True, autopct="%1.1f%%", explode=explode)
plt.title('Sample Pie Chart')
```

```python
## Practical Example for Data Visualization using Energy Production Data

import pandas as pd

production_data_df = pd.read_csv('Energy_Production_Dataset.csv')
production_data_df.head(5)
```

```python
## Total Sales in each month

monthly_production = production_data_df.groupby('Month_Name')['Production'].sum()
monthly_production
```

```python
## To create a bar chart for Monthly Production Data

plt.figure(figsize=(12,3))
monthly_production.plot(kind='bar', color='teal')
plt.title('Monthly Production Data')
```

```python
## To create a pie chart for seasonal production

seasonal_production_df = production_data_df.groupby(['Season'])['Production'].sum()
seasonal_production_df
```

```python
## To create the pie chart

explode = [0.2,0,0,0]
seasonal_production_df.plot(kind='pie', autopct="%1.1f%%",shadow=True, explode=explode)
plt.title('Energy Produced in Different Seasons')
```

```python
## First 5 records of the DataFrame
production_data_df.head(5)
```

```python
## To create a line chart to see the production trend over time ('Date')

production_trend = production_data_df.groupby(['Date'])['Production'].sum()
production_trend.head(5)
```

```python
## reset_index() resets the column name "Production" after aggregating the data

production_trend = production_data_df.groupby(['Date'])['Production'].sum().reset_index()
production_trend.head(5)
```

```python
## Approach 1 for creating a trend line

production_trend.plot(kind='line', xlabel='Row Number for the Date', ylabel='Energy Produced', title='Energy Production Trend over Time', color='skyblue', figsize=(20,7))
## In this graph, in X axis, Row index is considered instead of Date. But Indicating the Date is most appropriate thing while creating the trendlines.
## This was done in the 2nd approach below
```

```python
## Approach 2 for creating the trend line

plt.figure(figsize=(20,7))
plt.plot(production_trend['Date'], production_trend['Production'],color='green')
plt.xlabel('Date')
plt.ylabel('Energy Produced')
plt.title('Energy Production Trend over Time')
```
