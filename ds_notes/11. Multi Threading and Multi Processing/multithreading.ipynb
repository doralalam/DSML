#### Multi Threading

```python
import threading
import time
```

```python
## Create 2 different functions

def print_number():
    for i in range(5):
        print(f'Number: {i}')

def print_letter():
    for letter in 'abcde':
        print(f'Letter: {letter}')


start_time = time.time()
print_number()
print_letter()
time_to_finish = start_time-time.time()
print(f'Time taken to finish the execution: {time_to_finish}')
```

```python
## To observe some significant time difference, let's use timer

def print_number():
    for i in range(5):
        ## sleep for 2 seconds
        time.sleep(2)
        print(f'Number: {i}')

def print_letter():
    for letter in 'abcde':
        ## sleep for 2 seconds
        time.sleep(2)
        print(f'Letter: {letter}')


start_time = time.time()
print_number()
print_letter()
time_to_finish = time.time()-start_time
print(f'Time taken to finish the execution: {time_to_finish}')

## Let's assume, the time used to sleep here is used for accessing some IO peripherals, 20 seconds of time for execution is very significant
## In this case, multi-threading process is very handy, which helps 2 parts of the program to run concurrently with time efficieny
```

```python
def print_number():
    for i in range(5):
        ## sleep for 2 seconds
        time.sleep(2)
        print(f'Number: {i}')

def print_letter():
    for letter in 'abcde':
        ## sleep for 2 seconds
        time.sleep(2)
        print(f'Letter: {letter}')

## Create 2 threads
t1 = threading.Thread(target=print_number)
t2 = threading.Thread(target=print_letter)

start_time = time.time()

## Start the threads
t1.start()
t2.start()

## Wait for both the threads to complete their executions
t1.join()
t2.join()

time_to_finish = time.time()-start_time
print(f'Time taken to finish the execution: {time_to_finish}')

## With this code, both the functions run concurrently and time execution will be decreased to half
```
