```python
## To create threads
import threading
## To creat HTML get requests
import requests
## To create BeautifulSoup object to parse the response using HTML parse
from bs4 import BeautifulSoup


def fetch_content(url):
    ## Send HTML get request to the URL
    response = requests.get(url)
    ## To parse the content in response using HTML parser
    soup = BeautifulSoup(response.content, 'html.parser')
    print(f'Fetched {len(soup.text)} words from {url}')

## List of URLs to fetch the content
urls = [
    'https://www.tataaia.com',
    'https://www.policybazaar.com'
]

## List of dynamic threads to start and join the threads dynamically
threads = []

for url in urls:
    ## Create a thread and pass the arguments
    thread = threading.Thread(target=fetch_content, args=[url])
    ## thread = threading.Thread(target=fetch_content, args=(url,))
    ## Here, the URL argument must be given as either list or tuple
    ## In order to give the arg as tuple, we must use comma, otherwise, the thread will interpret the URL string as iterable and iterates letter by letter which gives awkwards response
    ## To append each thread in threads list
    threads.append(thread)
    ## To start the threads
    thread.start()

## To join the threads using thread list. i.e., next task will be done only after all the threads completes their respective executions
for thread in threads:
    thread.join()

print('All the pages fetched successfully!')
```
