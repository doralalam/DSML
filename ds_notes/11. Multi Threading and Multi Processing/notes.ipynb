#### Program, Process and Thread

1. Program:

    A Program is a sequence of instructions written in any programming language. Ex: Python, C, C++, Java

    Example: MS Paints is a windows application (with .exe file), which is also known as Program, that helps us to draw some painting

2. Process:

    A process is an instance of a program that is being executed

    Example: When we double click on the MS Paint app icon on the desktop, it will open the application which is known as process

    1. Multiple processes use separate memory space. Each process will have separate ID

    2. So, one process cannot corrupt the other process

    3. The disadvantage with multiple processes is that, if a process requires IO device related operations, then switching between these processes will increase the execution time

3. Thread:

    A thread is a unit of execution with in the process

    Example: In the MS Paints application, if we open it, the process will begin and then we will start drawinig by selecting a rectangle, or a circle etc., which is known as thread

    Selecting a rectangle is a unit of execution within the process which is known as Thread

    Threads are of two types:

    1. Single Threaded Process

    2. Multi Threaded Process
