#### Probability Distribution Function

1. Probability Distribution Function is a broad term that explains how probability of a random variable is distributed among its possible values.

    Example: 

                1. PMF for discrete

                2. PDF for continuous

                3. CDF for both

#### Types of Probability Distribution Function

1. Probability Mass Function (PMF):

    - Used for discrete random variables

    - Used when X takes countable values (0,1,2,....)

2. Probability Density Function (PDF):

    - Used for continuous random variables

    - Used when X takes uncountable infinite values

3. Cumulative Distribution Function (CDF):

    - Used for both discrete and continuous random variables
