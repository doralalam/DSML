#### Flask

Flask is a web framework created by using Python programming language

#### 1. Web Server Gateway Interface

1. In general, after developing an application, we deploy that application into any of cloud service such as AWS EC2, Azure App, Apache, GCP, IIS etc.

2. So, whenever an end user tries to access the application deployed in the cloud, the request was first sent from 

        end_user -----> web_server (AWS EC2) -----> web_application (developed using Flask Framework) -----> web_server ----> end_user
                                                |
                                                |
                                                V
                                    Procotocol used between web_server
                                    and web_application for communication is
                                    Web Server Gateway Interface

#### 2. Jinja2 Template Engine

1. Jinja2 Template Engine is a web template engine

2. A web template engine will combine any web_page with a data_source (SQL DB, CSV Sheet, MongoDB, ML Model) which is helpful to create dynamic web_pages

For Example:

    If we want to build an application which classifies an image into whether a cat or dog using the ML Model, we will first upload the image in a web_page which then get's passed to the data_source (ML Model) using Jinja2 template engine.

    Then the model, will classify that image and then gives the response to the web_page

    This is how, we can build a dynamic web_page
